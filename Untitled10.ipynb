```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib

# Dataset
data = {
    "joke_level": [9,2,8,3,7,1,9,4],
    "lateness": [3,1,9,4,6,2,8,3],
    "eating_level": [4,7,5,10,3,8,6,9],
    "friend_type": ["Clown","Mom","Gremlin","Foodie","Clown","Mom","Gremlin","Foodie"]
}

df = pd.DataFrame(data)

# Encode friend_type
df["friend_type_code"] = df["friend_type"].astype("category").cat.codes

X = df[["joke_level", "lateness", "eating_level"]]
y = df["friend_type_code"]

# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Model
model = RandomForestClassifier()
model.fit(X_train, y_train)

# Evaluation
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Save the model
joblib.dump(model, "friend_model.pkl")
print("Model saved as friend_model.pkl")
```
